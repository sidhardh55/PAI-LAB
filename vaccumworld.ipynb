```python
# Vacuum Cleaner Simulation with Suck, Right, Left actions

# --- Input ---
num_rooms = int(input("Enter number of rooms: "))

rooms = []
room_status = {}

for i in range(num_rooms):
    name = input(f"Enter name of Room {i+1}: ").strip()
    status = input(f"Enter status of {name} (Clean/Dirty): ").capitalize()
    rooms.append(name)
    room_status[name] = status

start_location = input(f"Enter starting location (choose from {rooms}): ").strip()

# --- Agent Simulation ---
current_index = rooms.index(start_location)
steps = 0

print("\nInitial State:", room_status)

while "Dirty" in room_status.values():
    steps += 1
    current_room = rooms[current_index]
    status = room_status[current_room]

    print(f"\nStep {steps}:")
    print("Agent Location:", current_room)
    print("Room Status:", status)

    # Perform action
    if status == "Dirty":
        action = "Suck"
        room_status[current_room] = "Clean"
    else:
        # Move right if not last room, else move left
        if current_index < num_rooms - 1:
            action = "Right"
            current_index += 1
        else:
            action = "Left"
            current_index -= 1

    print("Action:", action)
    print("Updated Rooms:", room_status)

print("\nFinal State:", room_status)
```
