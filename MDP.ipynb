```python
import numpy as np

class MDP:
    def __init__(self, states, actions, transition_prob, reward_function, gamma=0.9):
        self.states = states
        self.actions = actions
        self.transition_prob = transition_prob
        self.reward_function = reward_function
        self.gamma = gamma

    def value_iteration(self, max_iterations=3):
        V = {s: 0 for s in self.states}
        for iteration in range(max_iterations):
            print(f"Iteration {iteration + 1}:")
            for state in self.states:
                V[state] = max([sum(self.transition_prob(state, a, next_state) * 
                                    (self.reward_function(state, a, next_state) + self.gamma * V[next_state]) 
                                    for next_state in self.states) for a in self.actions])
                print(f"State {state} = {V[state]:.2f}")
        return V

    def policy_iteration(self, max_iterations=3):
        policy = {s: np.random.choice(self.actions) for s in self.states}
        V = {s: 0 for s in self.states}
        for iteration in range(max_iterations):
            print(f"\nPolicy Iteration {iteration + 1}:")
            for state in self.states:
                V[state] = sum(self.transition_prob(state, policy[state], next_state) * 
                               (self.reward_function(state, policy[state], next_state) + self.gamma * V[next_state])
                               for next_state in self.states)
                print(f"State: {state}, Action: {policy[state]}, Value: {V[state]:.2f}")
            policy_stable = True
            for state in self.states:
                action_values = [sum(self.transition_prob(state, a, next_state) * 
                                     (self.reward_function(state, a, next_state) + self.gamma * V[next_state]) 
                                     for next_state in self.states) for a in self.actions]
                best_action = self.actions[np.argmax(action_values)]
                if best_action != policy[state]:
                    policy[state] = best_action
                    policy_stable = False
            if policy_stable: break
        return V

# Setup for a 3x2 rectangular matrix (3 rows, 2 columns)
states = [(0, 0), (0, 1), (1, 0), (1, 1), (2, 0), (2, 1)]  # 3x2 grid
actions = ['UP', 'DOWN', 'LEFT', 'RIGHT']

# Transition function (probability of moving from one state to another)
transition_prob = lambda s, a, ns: 0.7 if s == ns else 0.3

# Reward function (reward when moving from one state to another)
reward_function = lambda s, a, ns: -2.71 if s == (0, 0) else 0  # Example reward function

mdp = MDP(states, actions, transition_prob, reward_function)

# Run Value Iteration
print("Value Iteration:")
V = mdp.value_iteration()

# Run Policy Iteration
print("\nPolicy Iteration:")
V = mdp.policy_iteration()

# Final Values after Policy Iteration
print("\nFinal Values after Policy Iteration:")
for s in states:
    print(f"State: {s}, Value: {V[s]:.2f}")
```
